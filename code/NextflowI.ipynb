# Hola munto en nextflow


```python
%%writefile sayhello.nf

nextflow.enable.dsl=2
process sayHello {
    input:
        val cheers

    output:
        stdout emit: verbiage

    script:
    """
    echo -n $cheers
    """
}

workflow {
    things = channel.of('Hola Mundo!', 'Hello world!', 'Bazinga!1')
    sayHello(things)
    sayHello.out.verbiage.view()
}
```

```python
! nextflow run /content/sayhello.nf
```

```python
%%writefile ejemplo1.nf 

// Codigo nextflow
nextflow.enable.dsl=2

process NUM_LINEAS {
    input:
    path read

    script:
    """
    printf '${read} '
    gunzip -c ${read} | wc -l
    """
}

reads_ch = Channel.fromPath( '/content/wengan_demo/ecoli/reads/EC*.fastq.gz')

workflow {
  NUM_LINEAS(reads_ch)
}
```
